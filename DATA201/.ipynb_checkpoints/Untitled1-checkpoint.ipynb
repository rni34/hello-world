```r
get_out_nutrition <- function(nodes){
    label <- nodes %>%
    html_node('.nutrition')
    value <- nodes %>%
    html_node()
    html_text()
    nutr_dataframe <- tibble(nutrition_info = label,
                            nutrition_value = value)
    return(nutr_dataframe)
}
```

```r
aubergine %>%
    html_node('.nutrition') %>%
    html_nodes('li') %>%
    map_df(get_out_nutrition)
```
